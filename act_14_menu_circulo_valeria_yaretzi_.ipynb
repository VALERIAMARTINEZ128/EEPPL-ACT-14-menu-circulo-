* El círculo y la esfera son dos de las figuras geométricas
* más perfectas y estudiadas en la historia de las matemáticas.
* Desde la antigüedad, estas formas han fascinado a matemáticos,
* astrónomos y filósofos por su simetría y presencia en la naturaleza.
* En la era digital, comprender cómo calcular sus propiedades básicas
* —área, perímetro y volumen— se ha convertido en un ejercicio
* fundamental para los estudiantes de programación, especialmente
* aquellos que inician su camino en el desarrollo de software

Área del Círculo Fórmula: A = π × r²
Perímetro (Circunferencia) del Círculo Fórmula: P = 2 × π × r o P = π × d
Volumen de la Esfera Fórmula: V = (4/3) × π × r³

```python
# ==========================================
# 📐 CALCULADORA GEOMÉTRICA BÁSICA 📐
# Elaborado por Angela Michelle Correa Lopez
# Fórmulas utilizadas:
# 🔵 Área del círculo: A = π * r²
# 🟢 Perímetro del círculo: P = 2 * π * r
# 🌐 Volumen de la esfera: V = (4/3) * π * r³
# ==========================================

import math

def areacirculo():
    print("\n🔵 CALCULANDO EL ÁREA DEL CÍRCULO 🔵")
    radio = float(input("📏 Ingresa el radio del círculo: "))
    area = math.pi * radio ** 2
    print(f"✨ Resultado: El área es {area:.2f} unidades²")

def perimetrocirculo():
    print("\n🟢 CALCULANDO EL PERÍMETRO DEL CÍRCULO 🟢")
    radio = float(input("📏 Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✨ Resultado: El perímetro es {perimetro:.2f} unidades")

def volumenesfera():
    print("\n🌐 CALCULANDO EL VOLUMEN DE LA ESFERA 🌐")
    radio = float(input("📏 Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * radio ** 3
    print(f"✨ Resultado: El volumen es {volumen:.2f} unidades³")

def menu():
    while True:
        print("\n" + "="*45)
        print("🧮 MENÚ PRINCIPAL - GEOMETRÍA BÁSICA 🧮")
        print("="*45)
        print("1️⃣ Calcular área del círculo")
        print("2️⃣ Calcular perímetro del círculo")
        print("3️⃣ Calcular volumen de la esfera")
        print("4️⃣ Salir del programa")
        print("="*45)

        opcion = input("👉 Elige una opción (1-4): ")

        if opcion == "1":
            areacirculo()
        elif opcion == "2":
            perimetrocirculo()
        elif opcion == "3":
            volumenesfera()
        elif opcion == "4":
            print("\n👋 ¡Programa finalizado! Gracias por calcular. 🎓✨")
            break
        else:
            print("⚠️ Opción no reconocida. Por favor, ingresa 1, 2, 3 o 4.")

# Punto de inicio del programa
if __name__ == "__main__":
    menu()
```
